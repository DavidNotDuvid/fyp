```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors
import numpy as np
import os
import seaborn as sb
```

```python
cwd = os.getcwd()
print(cwd)
df = pd.read_csv(cwd+"/src/experiments/12/time0")
```

```python
df.head()
```

```python
df = df[(df != 0).all(1)]
```

```python
df.describe(include = 'all')
```

```python
df_2 = df[['centroid-0','centroid-1']]
```

df_2

```python
df_2.head()
```

```python
df_2 = df_2.reset_index(drop=True)
```

```python
X= df_2.to_numpy()
```

```python
hi = sb.scatterplot(x= 'centroid-0', y = 'centroid-1', data =df_2,palette = "hsv")
```

```python
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=100).fit(X)
Y = clustering.labels_
```

```python
df_2['labels']=Y
```

```python
hi = sb.scatterplot(x= 'centroid-0', y = 'centroid-1', data =df_2, hue = 'labels',palette = "hsv")
```
